```python
import pandas as pd

def etree2df(root):
    column_names = []
    for i in range(len(root.getchildren()[0].getchildren())):
        column_names.append(root.getchildren()[0].getchildren()[i].tag)
    
    rows = []
    for j in range(len(root.getchildren())):
        obj = root.getchildren()[j].getchildren()
        texts = [obj[k].text for k in range(len(column_names))]
        row = dict(zip(column_names, texts))
        rows.append(row)
    
    # Create the DataFrame using pd.DataFrame directly from the list of rows
    xml_frame = pd.DataFrame(rows, columns=column_names)
    return xml_frame

# Call the function with your XML root element
etree2df(root)
```

```python
# Extract all rows from the XML data
all_rows = []
for book in root.findall('Book'):
    row = [child.text for child in book.getchildren()]
    all_rows.append(row)

# Display the extracted rows
for row in all_rows:
    print(row)
```

```python
# The Format HDF5( Hierarchical Data Format )
# Advantages: Hierarchical Structure,  Efficient Storage, Support for Multiple Data Types,Cross-Platform Compatibility, Random Access

import pandas as pd
import numpy as np
from pandas.io.pytables import HDFStore
# HDFStore is a class used to handle storage in HDF5 files, specifically through the PyTables library.

# Create the first DataFrame
frame = pd.DataFrame(np.arange(16).reshape(4, 4), 
                     index=['white', 'black', 'red', 'blue'], 
                     columns=['up', 'down', 'right', 'left'])

# Create the second DataFrame
data = {
    'up': [0, 2, 4, 6],
    'down': [0.5, 2.5, 4.5, 6.5],
    'right': [1, 3, 5, 7],
    'left': [1.5, 3.5, 5.5, 7.5]
}
index = ['white', 'black', 'red', 'blue']
frame2 = pd.DataFrame(data, index=index)


# HDFStore is used to create an HDF5 file named mydata.h5.
# HDFStore allows you to store multiple objects (DataFrames) within the same HDF5 file, similar to saving multiple sheets in an Excel file.

# Create an HDF5 file using HDFStore
with HDFStore('data.h5') as store:
    # Storing multiple DataFrames within the same HDF5 file
    store['obj1'] = frame
    store['obj2'] = frame2

# Save the first DataFrame to an HDF5 file using to_hdf
frame.to_hdf('data.h5', key='dataset1', mode='w')

# Save the second DataFrame to the same HDF5 file using to_hdf
frame2.to_hdf('data.h5', key='dataset2', mode='a')  # Use mode='a' to append

# Read the data back from the HDF5 file
loaded_data1 = pd.read_hdf('data.h5', key='dataset1')
loaded_data2 = pd.read_hdf('data.h5', key='dataset2')

# Display the loaded data
print("Loaded Data from dataset1:")
print(loaded_data1)

print("\nLoaded Data from dataset2:")
print(loaded_data2)
```

```python
import pandas as pd
import numpy as np
from pandas.io.pytables import HDFStore

# Create the first DataFrame with one structure
frame1 = pd.DataFrame({
    'A': [1, 2, 3],
    'B': [4, 5, 6],
    'C': [7, 8, 9]
})

# Create a second DataFrame with a different structure
frame2 = pd.DataFrame({
    'X': ['foo', 'bar', 'baz'],
    'Y': [10.5, 20.5, 30.5],
    'Z': ['apple', 'banana', 'cherry'],
    'W': [True, False, True]
})
# HDFStore is a pandas class that provides a way to interact with HDF5 files.
# Use HDFStore to save both DataFrames in the same HDF5 file
with HDFStore('different_structures.h5') as store:
    store['frame1'] = frame1
    store['frame2'] = frame2

# Read the DataFrames back from the HDF5 file
loaded_frame1 = pd.read_hdf('different_structures.h5', key='frame1')
loaded_frame2 = pd.read_hdf('different_structures.h5', key='frame2')

# Display the loaded DataFrames to verify they were saved and loaded correctly
print("Loaded DataFrame 1:")
print(loaded_frame1)

print("\nLoaded DataFrame 2:")
print(loaded_frame2)
```

```python
import pandas as pd
from sqlalchemy import create_engine

# Create a DataFrame
frame = pd.DataFrame(np.arange(20).reshape(4, 5), columns=['white', 'red', 'blue', 'black', 'green'])
print("Original DataFrame:")
print(frame)

# Setting up the SQLite Database Connection
engine = create_engine('sqlite:///foo.db')

# Convert the DataFrame into a table  colorsTable within the database
frame.to_sql('colorsTable', engine, if_exists='replace', index=False) 

# Reading the table back from the database using an SQL query
frameLoaded = pd.read_sql_query('SELECT * FROM colorsTable', engine)

print("\nLoaded DataFrame from SQLite Database:")
print(frameLoaded)
```

```python
import pymongo
from pymongo import MongoClient
import pandas as pd
import numpy as np
import json


# client is a connection object that represents the MongoDB server itself. It is used to establish a connection to your MongoDB server,
# which is listening on a specific host (localhost) and on the default port (27017).
client = MongoClient('localhost', 27017)

# Point to a specific database and collection
db = client.mydatabase 
# db represents a specific database within the MongoDB server. It acts as a container for your data collections.
# MongoDB databases are used to organize your collections in a logical way.mydatabase is the name of the database you're connecting to.
# If it doesn't exist, MongoDB will create it when you first store data in it.

# collection is similar to a table in relational databases. It contains a group of documents (records) within a specific database.
collection = db.mycollection

# Create a DataFrame
frame = pd.DataFrame(np.arange(20).reshape(4, 5), columns=['white', 'red', 'blue', 'black', 'green'])
print(frame)

# Convert DataFrame to JSON format suitable for MongoDB insertion
records = json.loads(frame.T.to_json()).values()
print("records")
print(records,"\n")

# Insert records into MongoDB collection
collection.insert_many(records)

# Fetch the inserted data from the collection
cursor = collection.find()
dataframe = pd.DataFrame(list(cursor))

# Remove the '_id' field from the DataFrame
if '_id' in dataframe.columns:
    del dataframe['_id']

# Display the resulting DataFrame
print("Display the resulting DataFrame")
print(dataframe)
```

```python
import pymongo
from pymongo import MongoClient
import pandas as pd
import numpy as np
import json

# Establish a connection to the MongoDB server
client = MongoClient('localhost', 27017)

# Point to a specific database and collection
db = client.mydatabase 
collection = db.mycollection

# Create a DataFrame
frame = pd.DataFrame(np.arange(20).reshape(4, 5), columns=['white', 'red', 'blue', 'black', 'green'])
print("Original DataFrame:")
print(frame)

# Convert DataFrame to a list of dictionaries suitable for MongoDB insertion
records = frame.to_dict(orient='records')
print("\nRecords to be inserted into MongoDB:")
print(records)

# Insert records into MongoDB collection
collection.insert_many(records)

# Fetch the inserted data from the collection
cursor = collection.find()
dataframe = pd.DataFrame(list(cursor))

# Remove the '_id' field from the DataFrame
if '_id' in dataframe.columns:
    del dataframe['_id']

# Display the resulting DataFrame
print("\nDisplay the resulting DataFrame from MongoDB:")
print(dataframe)
```

```python
from pymongo import MongoClient
import pandas as pd
import numpy as np
import json

# Try to connect to MongoDB server with a timeout of 5 seconds
try:
    client = MongoClient('localhost', 27017, serverSelectionTimeoutMS=5000)
    # Attempt to get server information to check if the server is running
    client.server_info()
    print("Connected to MongoDB server successfully.")
    
    # Point to a specific database and collection
    db = client.mydatabase
    collection = db.mycollection

    # Create a DataFrame
    frame = pd.DataFrame(np.arange(20).reshape(4, 5), columns=['white', 'red', 'blue', 'black', 'green'])
    print("Original DataFrame:")
    print(frame)

    # Convert DataFrame to a list of dictionaries suitable for MongoDB insertion
    records = frame.to_dict(orient='records')
    print("\nRecords to be inserted into MongoDB:")
    print(records)

    # Insert records into MongoDB collection
    collection.insert_many(records)

    # Fetch the inserted data from the collection
    cursor = collection.find()
    dataframe = pd.DataFrame(list(cursor))

    # Remove the '_id' field from the DataFrame
    if '_id' in dataframe.columns:
        del dataframe['_id']

    # Display the resulting DataFrame
    print("\nDisplay the resulting DataFrame from MongoDB:")
    print(dataframe)

except Exception as e:
    print("Could not connect to MongoDB server:", e)
```
